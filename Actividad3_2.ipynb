```python
data_num_2021 = gastos_costos_2021[['TOTAL SAT']]
data_num_2021
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data_num_2021.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

# Desviación estándar

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data_num_2021
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data_num_2021[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= data_num_2021[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean) 
plt.title("Outliers de Gastos y Costos 2021")
plt.show() #dibujamos el diagrama
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

cualitativos =gastos_costos_2021[['FECHA', 'TIPO GASTO', 'TC']]
Datos_limpios_2021 = pd.concat([cualitativos, data_clean], axis=1)
Datos_limpios_2021
```

```python
#Convertir DataFrame a CSV
Datos_limpios_2021.to_csv("Datos_limpios.csv")
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data_num_2021

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data_num_2021[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= data_num_2021[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data4_iqr) 
plt.title("Outliers de Gastos y Costos 2021")
plt.show() #dibujamos el diagrama
```
